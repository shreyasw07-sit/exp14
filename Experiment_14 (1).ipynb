```python
import pandas as pd
import numpy as np
data = { 'Product': ['Laptop','Mobile', 'Tablet', 'Headphones', 'Smartwatch', 'Camera'], 'Price': [55000, 20000, 30000, 2000, 8000, 45000],'Units_Sold':[25,60,35,80,50,15]}
df = pd.DataFrame(data)
print(df)
```

```python
# Data Binning
bins1 = [0, 10000, 30000, 60000]
labels1 = ['Low','Medium', 'High']
df['Price_Category'] = pd.cut(df['Price'], bins=bins1, labels=labels1)
print(df)
```

```python
# Binning Units Sold
bins2 = [0, 30, 60, 100]
labels2 = ['Low Sales','Medium Sales', 'High Sales']
df['Sales_Category'] = pd.cut(df['Units_Sold'], bins=bins2, labels=labels2)
print(df)
```

```python
# Data Formatting
# Convert Data Type
print(df.dtypes)
df['Units_Sold'] = df['Units_Sold'].astype(float)
print(df.dtypes)
```

```python
df
```

```python
# Convert Text to Uppercase
df['Product'] = df['Product'].str.upper()
print(df)
```

```python
# Rounding Values: rounding values like for decimals it would round it to whole number
df['Price'] = df['Price'].round(2)
print(df)
```

```python
# Additional Operations
# Sort Data
sorted_df = df.sort_values(by= 'Price')
print(sorted_df)
```

```python
# Sort data in descending order
sorted_df1 = df.sort_values(by= 'Price', ascending=False)
print(sorted_df1)
```

```python
# Display Unique Values
print(df['Price_Category'].unique())
```

```python
# Dataset: Food Delivery Orders
import pandas as pd
import numpy as np

data = {'Order_ID':['O1','O2','O3','O4','O5','O6','O7','O8'], 'Order_Value':[250,450,120,600,300,150,500,220], 'Delivery_Time':[30,45,25,60,35,20,50,28],'Distance_km':[5,8,3,10,6,2,9,4]}
df = pd.DataFrame(data)
print(df)
```

```python
# Binning Order Value
bins = [0, 200,400, 700]
labels = ['Low','Medium', 'High']
df['Order_Value_Category'] = pd.cut(df['Order_Value'], bins=bins, labels= labels)
print(df)
```

```python
# Binning Delivery Time
bins1 = [0, 25, 40, 60]
labels1= ['Late', 'On Time', 'Early']
df['Delivery_Time_Category'] = pd.cut(df['Delivery_Time'], bins=bins1, labels= labels1)
print(df)
```

```python
bins2 = [ 0, 4, 8, 10]
labels2= ['Short', 'Medium', 'High']
df['Distance_km_Category'] = pd.cut(df['Distance_km'], bins=bins2, labels= labels2)
print(df)
```

```python
print(df.dtypes)
df['Delivery_Time'] = df['Delivery_Time'].astype(float)
print(df.dtypes)
```

```python
df['Order_Value_Category'] = df['Order_Value_Category'].str.upper()
print(df)
```

```python
sorted_df = df.sort_values(by= 'Distance_km')
print(sorted_df)
```

```python
sorted_df = df.sort_values(by= 'Order_Value')
print(sorted_df)
```
